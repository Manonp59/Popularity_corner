```python
import requests

def get_id(name, api_key):
    base_url = 'https://api.themoviedb.org/3/search/movie'
    url = f'{base_url}?api_key={api_key}&query={name}'
    response = requests.get(url)
    if response.status_code == 200:
        search_results = response.json()
        if search_results['total_results'] > 0:
            # Prendre le premier résultat de la liste (vous pouvez afficher tous les résultats si nécessaire)
            first_result = search_results['results'][0]
            movie_id = first_result['id']
            return movie_id
        else:
            return f"Aucun résultat trouvé pour '{name}'."
    else:
        return "Erreur lors de la requête API 1 ."

def get_budget(movie_id, api_key):
    base_url = 'https://api.themoviedb.org/3/movie/'
    url = f'{base_url}{movie_id}?api_key={api_key}'

    # Effectuer la requête
    response = requests.get(url)

    # Analyser la réponse JSON
    if response.status_code == 200:
        movie_data = response.json()
        budget = movie_data.get('budget')
        return budget
    else:
        return "Erreur lors de la requête API 2."
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv('API_KEY_BUDGET')

for index, row in df.iterrows():
    title = row['title']
    budget = get_budget(get_id(title, api_key), api_key)
    df.at[index, 'budget'] = budget
```

```python
import pandas as pd

df_meteo = pd.read_csv('weekly_average_temperatures.csv')
```

```python

import dateparser

# Convertir la colonne 'release_date' au format souhaité (jour mois année)
df['release_date'] = df['release_date'].apply(lambda x: dateparser.parse(x).date() if isinstance(x, object) and dateparser.parse(x) else None)
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
import pandas as pd

# Assurez-vous que les colonnes "Start Date" et "End Date" de df_meteo sont au format datetime
df_meteo['Start Date'] = pd.to_datetime(df_meteo['Start Date'])
df_meteo['End Date'] = pd.to_datetime(df_meteo['End Date'])

# Ajouter une colonne "Temperature Moyenne" à df_films
df['Temperature Moyenne'] = None

villes = ['Lille', 'Bordeaux', 'Lyon', 'Paris']
for ville in villes:
    df[ville] = None

# Parcourir chaque date de sortie dans df_films
for index, row in df.iterrows():
    release_date = row['release_date']
    
    # Rechercher la semaine correspondante dans df_meteo
    matching_week = df_meteo[(df_meteo['Start Date'] <= release_date) & (df_meteo['End Date'] >= release_date)]
    
    if not matching_week.empty:
        # Récupérer la température moyenne pour cette semaine
        temperature_moyenne = matching_week.iloc[0]['Moyenne']
        
        # Ajouter la température moyenne à la colonne correspondante dans df_films
        df.at[index, 'Temperature Moyenne'] = temperature_moyenne
        
        for ville in villes:
            temperature_ville = matching_week.iloc[0][ville]
            df.at[index, ville] = temperature_ville

```

```python
import re

def format_date(date_list):
    formatted_dates = []
    for date_str in date_list:
        # Try to match date range format
        match_range = re.match(r'(\d{1,2}) au (\d{1,2}) ([a-zéû]+) (\d{4})', date_str, re.I)
        if match_range:
            day_start, day_end, month_str, year = match_range.groups()
            day_start = int(day_start)
            day_end = int(day_end)
            months = {'janvier': 1, 'février': 2, 'mars': 3, 'avril': 4, 'mai': 5, 'juin': 6,
                      'juillet': 7, 'août': 8, 'septembre': 9, 'octobre': 10, 'novembre': 11, 'décembre': 12}
            month = months[month_str.lower()]
            year = int(year)
            formatted_date = f'{year}-{month:02d}-{day_start:02d}'
            formatted_dates.append(formatted_date)
        else:
            # Try to match single date format
            match_range_1 = re.match(r'(\d{1,2}) ([a-zéû]+) (?:au|jusqu\'au) (\d{1,2}) ([a-zéû]+) (\d{4})', date_str, re.I)
            match_range_2 = re.match(r'(\d{1,2}) ([a-zéû]+) (\d{4})? (?:au|jusqu\'au) (\d{1,2}) ([a-zéû]+) (\d{4})', date_str, re.I)
            if match_range_1:
                day_start, month_start, day_end, month_end, year = match_range_1.groups()
                day_start, day_end = int(day_start), int(day_end)
                months = {'janvier': 1, 'février': 2, 'mars': 3, 'avril': 4, 'mai': 5, 'juin': 6,
                      'juillet': 7, 'août': 8, 'septembre': 9, 'octobre': 10, 'novembre': 11, 'décembre': 12}
                month_start = months.get(month_start.lower())
                month_end = months.get(month_end.lower())
                if month_start and month_end:
                    formatted_dates.append(f'{year}-{month_start:02d}-{day_start:02d}')
                    formatted_dates.append(f'{year}-{month_end:02d}-{day_end:02d}')
            elif match_range_2:
                    day_start, month_start, year_start, day_end, month_end, year_end = match_range_2.groups()
                    day_start, day_end = int(day_start), int(day_end)
                    months = {'janvier': 1, 'février': 2, 'mars': 3, 'avril': 4, 'mai': 5, 'juin': 6,
                      'juillet': 7, 'août': 8, 'septembre': 9, 'octobre': 10, 'novembre': 11, 'décembre': 12}
                    month_start = months.get(month_start.lower())
                    month_end = months.get(month_end.lower())
                    if not year_start:
                        year_start = year_end
                    if month_start and month_end:
                        formatted_dates.append(f'{year_start}-{month_start:02d}-{day_start:02d}')
                        formatted_dates.append(f'{year_end}-{month_end:02d}-{day_end:02d}')
            else:
                formatted_dates.append(date_str)  # Handle invalid date strings
    return formatted_dates

# Test avec la liste de dates que vous avez fournie
df['box_office_date'] = df['box_office_date'].apply(format_date)
```

```python
df['box_office_date'] = df['box_office_date'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
```

```python
# Convertir les colonnes de dates en objets datetime
df['box_office_date'] = pd.to_datetime(df['box_office_date'], errors='coerce')

# Fonction pour obtenir le box office définitif
def get_final_box_office(row):
    if pd.notnull(row['box_office_date']) and row['release_date'] <= row['box_office_date']:
        return row['box_office_first_week']
    else:
        return row['box_office_second_week']

# Appliquer la fonction pour obtenir la colonne 'box_office_final'
df['box_office_first_week'] = df.apply(get_final_box_office, axis=1)

# Supprimer les colonnes temporaires (box_office_first_date, box_office_first, box_office_second)
df.drop(['box_office_date', 'box_office_second_week'], axis=1, inplace=True)
```
